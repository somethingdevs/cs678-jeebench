# Encode Answers

```python
import pandas as pd

def encode_answers(df):
    # Define bins for numeric answers to categorize them
    bins = [-float('inf'), 0, 50, 100, float('inf')]
    labels = ['Negative', '0-50', '51-100', '100+']

    # Convert all answers to strings to standardize input
    df.loc[:, 'answer'] = df['answer'].astype(str)

    # Normalize text answers, treating 'none' and 'None' as a special category
    df.loc[:, 'answer_normalized'] = df['answer'].apply(lambda x: 'None' if x.lower() == 'none' else x)

    # Safely convert answers to numeric, or set as NaN if conversion fails
    df.loc[:, 'numeric'] = pd.to_numeric(df['answer_normalized'], errors='coerce')

    # Apply bins to numeric answers and use predefined labels
    df.loc[:, 'binned_numeric'] = pd.cut(df['numeric'], bins=bins, labels=labels).astype(str)

    # Combine binned numeric answers with categorical answers
    df.loc[:, 'final_answer'] = df['binned_numeric'].where(df['binned_numeric'] != 'nan', df['answer_normalized'])

    # One-hot encode all answers
    encoded_answers = pd.get_dummies(df['final_answer'])
    return encoded_answers


# Encode the answers
train_encoded = encode_answers(train_data)
val_encoded = encode_answers(val_data)
test_encoded = encode_answers(test_data)


# Ensure both datasets have the same dummy columns
all_columns = set(train_encoded.columns) | set(val_encoded.columns) | set(test_encoded.columns)
train_encoded = train_encoded.reindex(columns=all_columns, fill_value=0)
val_encoded = val_encoded.reindex(columns=all_columns, fill_value=0)
test_encoded = test_encoded.reindex(columns=all_columns, fill_value=0)

# Display the encoded data for verification
print("Encoded Train Data:")
print(train_encoded.head())
print("\nEncoded Validation Data:")
print(val_encoded.head())
```

# Tokenize Questions

```python
from transformers import BertTokenizer

# Initialize the tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

def tokenize_questions(df):
    """Tokenize the questions and add tokenized data to the DataFrame."""
    # Tokenize and encode the questions for use with mBERT
    df['tokenized_data'] = df['question'].apply(lambda x: tokenizer.encode_plus(
        x,
        add_special_tokens=True,  # Add '[CLS]' at the start and '[SEP]' at the end
        max_length=512,          # Pad or truncate to a maximum length allowed by BERT
        truncation=True,
        padding='max_length',    # Pad to max_length
        return_tensors='pt'      # Return PyTorch tensors
    ))

    # Extract 'input_ids' and 'attention_mask' and assign them to new columns
    df['input_ids'] = df['tokenized_data'].apply(lambda x: x['input_ids'].squeeze(0))  # Remove the batch dimension
    df['attention_mask'] = df['tokenized_data'].apply(lambda x: x['attention_mask'].squeeze(0))

    return df

# Tokenize the questions in each dataset separately
train_data = tokenize_questions(train_data)
val_data = tokenize_questions(val_data)
test_data = tokenize_questions(test_data)

# Display the first few entries to verify tokenization
print("Tokenized Train Data:")
print(train_data[['question', 'input_ids', 'attention_mask']].head())

print("\nTokenized Validation Data:")
print(val_data[['question', 'input_ids', 'attention_mask']].head())

print("\nTokenized Test Data:")
print(test_data[['question', 'input_ids', 'attention_mask']].head())
```

# mBERT

```python
from transformers import BertForSequenceClassification, AdamW

# Load the pre-trained Multilingual BERT model with a sequence classification head
model = BertForSequenceClassification.from_pretrained(
    'bert-base-multilingual-cased',  # Base multilingual cased model
    num_labels=19,                   # The number of output labels. Adjust this based on the number of your classes
    output_attentions=False,         # Whether the model returns attentions weights.
    output_hidden_states=False,      # Whether the model returns all hidden-states.
)

# Move the model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
# Define the optimizer with weight decay fixed, except for bias and LayerNorm parameters
no_decay = ['bias', 'LayerNorm.weight']
optimizer_grouped_parameters = [
    {'params': [p for n, p in model.named_parameters() if not any(nd in n for nd in no_decay)], 'weight_decay': 0.01},
    {'params': [p for n, p in model.named_parameters() if any(nd in n for nd in no_decay)], 'weight_decay': 0.0}
]

optimizer = AdamW(optimizer_grouped_parameters, lr=1e-5)
```

```python
from transformers import get_linear_schedule_with_warmup

# Define the number of training steps
num_epochs = 10
num_training_steps = len(train_loader) * num_epochs

# Setup the scheduler
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=0,  # Usually around 10% of total training steps
    num_training_steps=num_training_steps
)
```

# Training Loop

```python
from torch.cuda.amp import GradScaler, autocast

def train_model(model, train_loader, val_loader, optimizer, scheduler, num_epochs, device, accumulation_steps=4):
    scaler = GradScaler()  # Initialize the gradient scaler for mixed precision training
    model.train()

    for epoch in range(num_epochs):
        total_train_loss = 0
        model.zero_grad()  # Reset gradients tensors

        for step, (batch_inputs, batch_masks, batch_labels) in enumerate(train_loader):
            batch_inputs, batch_masks, batch_labels = batch_inputs.to(device), batch_masks.to(device), batch_labels.to(device)

            with autocast():  # Mixed precision context
                outputs = model(input_ids=batch_inputs, attention_mask=batch_masks)
                logits = outputs.logits
                loss = torch.nn.functional.cross_entropy(logits, batch_labels) / accumulation_steps  # Scale loss

            scaler.scale(loss).backward()  # Scale the loss and perform backward pass with scaled gradients

            if (step + 1) % accumulation_steps == 0:
                scaler.step(optimizer)  # Update model parameters
                optimizer.zero_grad()  # Clear gradients
                scaler.update()  # Update the scale for next iteration
                scheduler.step()  # Update learning rate

            total_train_loss += loss.item() * accumulation_steps  # Unscaled loss for logging

        avg_train_loss = total_train_loss / len(train_loader)
        print(f"Epoch {epoch + 1}/{num_epochs} | Train Loss: {avg_train_loss}")

        validate_model(model, val_loader, device)

def validate_model(model, val_loader, device):
    # Switch model to evaluation mode
    model.eval()

    total_val_loss = 0
    total_val_accuracy = 0

    with torch.no_grad():
        for batch_inputs, batch_masks, batch_labels in val_loader:
            # Move batch data to the device
            batch_inputs, batch_masks, batch_labels = batch_inputs.to(device), batch_masks.to(device), batch_labels.to(device)

            # Forward pass with attention masks
            outputs = model(input_ids=batch_inputs, attention_mask=batch_masks)
            logits = outputs.logits

            # Calculate loss
            loss = torch.nn.functional.cross_entropy(logits, batch_labels)
            total_val_loss += loss.item()

            # Calculate accuracy
            preds = torch.argmax(logits, dim=1)
            if batch_labels.dim() > 1:  # Check if labels are one-hot encoded
                batch_labels = torch.argmax(batch_labels, dim=1)

            accuracy = (preds == batch_labels).float().mean()
            total_val_accuracy += accuracy.item()

    avg_val_loss = total_val_loss / len(val_loader)
    avg_val_accuracy = total_val_accuracy / len(val_loader)
    print(f"Validation Loss: {avg_val_loss} | Validation Accuracy: {avg_val_accuracy}")
```

```python
# Example call to train the model
train_model(model, train_loader, val_loader, optimizer, scheduler, num_epochs, device)
```

```python
def test_model(model, test_loader, device):
    model.eval()  # Set the model to evaluation mode
    total_test_loss = 0
    total_test_accuracy = 0

    with torch.no_grad():  # Turn off gradients for testing, saves memory and computations
        for batch_inputs, batch_masks, batch_labels in test_loader:
            batch_inputs, batch_masks, batch_labels = batch_inputs.to(device), batch_masks.to(device), batch_labels.to(device)

            outputs = model(input_ids=batch_inputs, attention_mask=batch_masks)
            logits = outputs.logits
            loss = torch.nn.functional.cross_entropy(logits, batch_labels)
            total_test_loss += loss.item()

            preds = torch.argmax(logits, dim=1)

            if batch_labels.dim() > 1:  # Check if labels are one-hot encoded
                batch_labels = torch.argmax(batch_labels, dim=1)
            accuracy = (preds == batch_labels).float().mean()
            total_test_accuracy += accuracy.item()

    avg_test_loss = total_test_loss / len(test_loader)
    avg_test_accuracy = total_test_accuracy / len(test_loader)
    print(f"Test Loss: {avg_test_loss} | Test Accuracy: {avg_test_accuracy}")

# Call the testing function
test_model(model, test_loader, device)
```

```python
def test_model_by_language(model, test_loader, device, language="all"):
    model.eval()  # Set the model to evaluation mode
    total_test_loss = 0
    total_test_accuracy = 0

    with torch.no_grad():  # Turn off gradients for testing, saves memory and computations
        for batch_inputs, batch_masks, batch_labels in test_loader:
            batch_inputs, batch_masks, batch_labels = batch_inputs.to(device), batch_masks.to(device), batch_labels.to(device)

            outputs = model(input_ids=batch_inputs, attention_mask=batch_masks)
            logits = outputs.logits
            loss = torch.nn.functional.cross_entropy(logits, batch_labels)
            total_test_loss += loss.item()

            preds = torch.argmax(logits, dim=1)

            if batch_labels.dim() > 1:  # Check if labels are one-hot encoded
                batch_labels = torch.argmax(batch_labels, dim=1)
            accuracy = (preds == batch_labels).float().mean()
            total_test_accuracy += accuracy.item()

    avg_test_loss = total_test_loss / len(test_loader)
    avg_test_accuracy = total_test_accuracy / len(test_loader)
    print(f"Language: {language}, Test Loss: {avg_test_loss}, Test Accuracy: {avg_test_accuracy}")
```

```python
def create_language_loader(test_data, language, tokenizer, batch_size=16):
    # Filter data by language
    lang_data = test_data[test_data['language'] == language]

    # Tokenize and prepare data
    tokenized_data = tokenize_questions(lang_data)

    # Encode answers, assuming this function originally returns one-hot encoded vectors
    encoded_answers = encode_answers(tokenized_data)
    # Convert one-hot encoded vectors to class indices
    labels = torch.argmax(torch.tensor(encoded_answers.values, dtype=torch.float), dim=1)

    # Create tensors for features and masks
    features = torch.stack(list(tokenized_data['input_ids'])).to(torch.long)
    masks = torch.stack(list(tokenized_data['attention_mask'])).to(torch.long)

    # Create TensorDataset and DataLoader
    dataset = TensorDataset(features, masks, labels)
    return DataLoader(dataset, batch_size=batch_size, shuffle=False)
```

```python
languages = test_data['language'].unique()  # Assuming 'language' column exists in your DataFrame
for lang in languages:
    lang_loader = create_language_loader(test_data, lang, tokenizer)
    test_model_by_language(model, lang_loader, device, language=lang)
```
